```python
from novel_agent import run_novel_agent
import os

if __name__ == "__main__":
    print("📖 歡迎使用小說創作 AI Agent")
    print("請輸入本章節目標，例如：")
    print("  主角前往星環之塔，並遇見一位神秘的學徒。")

    chapter_goal = input("\n請輸入章節目標：\n> ")

    filename = input("\n要將這章儲存為檔案嗎？（輸入檔名或留空跳過）\n> ")
    if filename:
        if not filename.endswith(".md"):
            filename += ".md"
        save_path = f"outputs/{filename}"
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
    else:
        save_path = None

    print("\n🧠 開始由 GPT 生成小說內容...\n")
    chapter_text = run_novel_agent(chapter_goal=chapter_goal, save_to=save_path)

    print("\n📄 生成完成，以下是本章內容：\n")
    print("=" * 40)
    print(chapter_text)
    print("=" * 40)

    if save_path:
        print(f"\n✅ 本章已儲存至：{save_path}")
```
