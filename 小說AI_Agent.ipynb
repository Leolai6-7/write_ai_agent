```python
import yaml

def save_yaml_file(name: str, content: dict):
    """將 YAML 字典內容寫入指定檔案名稱（自動加上 .yaml 副檔名）"""
    filename = f"{name}.yaml"
    with open(filename, 'w', encoding='utf-8') as f:
        yaml.dump(content, f, allow_unicode=True)
    print(f"✅ 檔案已儲存：{filename}")
```

```python
main_character = {
    "name": "伊澤·艾爾文",
    "age": 17,
    "gender": "男性",
    "race": "人類",
    "occupation": "魔導學徒",
    "personality": ["倔強", "極具求知慾", "不輕易相信他人"],
    "background": "伊澤出生於東大陸邊境的阿羅恩村，母親是草藥師，父親在他幼年時失踪。傳言父親曾是星環之塔的高階魔法師，後來與禁咒有關的實驗失敗而失踪。他從小就對魔法極為敏感。",
    "goals": ["找出父親失蹤的真相", "成為星環之塔的正式魔導士", "解開禁咒的本質與歷史真相"],
    "speaking_style": "語氣冷靜，言語簡潔直接，思考時喜歡自言自語。",
    "skills": ["初級元素魔法（火、水）", "古代語文辨識", "生存技能（草藥辨識、陷阱設置）"],
    "weaknesses": ["容易情緒內耗", "對禁咒與父親的話題過度執著"]
}

save_yaml_file("main_character", main_character)
```

```python
world_setting = {
    "continent": "艾爾特大陸",
    "eras": "第五紀元",
    "magic_system": {
        "source": "以靈脈為基礎，分五系：元素、精神、轉化、召喚、禁咒",
        "rules": [
            "魔力來自大地與星辰交會的節點",
            "使用魔法會削弱生命精華，需休養或借助靈晶恢復"
        ]
    },
    "races": ["人類", "精靈族", "矮人"],
    "geography": ["阿羅恩村", "星環之塔", "絕境深淵"]
}

save_yaml_file("world_setting", world_setting)
```

```python
from openai import OpenAI
import yaml
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")


# 初始化 client（API Key 會自動讀取 OPENAI_API_KEY 環境變數）
client = OpenAI()

def load_yaml(path):
    with open(path, 'r', encoding='utf-8') as f:
        return yaml.safe_load(f)

def build_prompt(character, world, objective):
    return f"""你是一位中文奇幻小說作家，正在撰寫一部關於「人類少年魔導學徒{character['name']}」的長篇小說。

請根據以下角色設定與章節目標，寫出故事的第一章開頭（大約 2~4 段，約 300~500 字），風格內斂、語句流暢、帶有些許神秘感，並保留後續劇情懸念。

【角色設定】
- 名字：{character['name']}，{character['age']} 歲，{character['race']}，{character['occupation']}
- 性格：{', '.join(character['personality'])}
- 背景：{character['background']}
- 目標：{', '.join(character['goals'])}

【世界觀】
- 時代：{world['eras']}
- 魔法系統：{world['magic_system']['source']}；規則：{', '.join(world['magic_system']['rules'])}
- 地點示例：{', '.join([p for p in world['geography']])}

【章節目標】
{objective}

請寫出第一章〈遺物之謎〉的內容，讓讀者進入{character['name']}的視角與情感中。
"""

def generate_chapter(character_path, world_path, objective):
    character = load_yaml(character_path)
    world = load_yaml(world_path)
    prompt = build_prompt(character, world, objective)

    # 使用 chat.completions 接口
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.8,
        max_tokens=1000
    )

    return response.choices[0].message.content.strip()
```

```python
chapter_text = generate_chapter(
    "main_character.yaml",
    "world_setting.yaml",
    "母親過世後，伊澤在整理遺物時，發現壁爐後方藏有一本發黃的手札，裡面提到他父親與某場禁咒實驗有關。"
)

print(chapter_text)
```
